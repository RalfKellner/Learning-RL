```python
import pandas as pd
import numpy as np
import itertools
from sklearn.preprocessing import StandardScaler
from datetime import datetime
import pickle
import matplotlib.pylab as plt
import random

def get_data():
    df = pd.read_csv('aapl_msi_sbux.csv')
    return df.values


#class for the agent for acting and learning
class DQNAgent():
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.epsilon = 0.90
        self.epsilon_decay = 0.99
        self.epsilon_min = 0.01
        self.gamma = 0.95
        self.model = LinearModel(state_size, action_size)
        
    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.choice(np.arange(0, self.action_size))
        else:
            pred = self.model.predict(state)
            return np.argmax(pred[0])
        
        
    def train(self, state, action, reward, next_state, done):
        if done:
            target = reward
        else:
            target = reward + self.gamma * np.amax(self.model.predict(next_state)[0])
            
        target_full = self.model.predict(state)
        target_full[0, action] = target
        
        self.model.sgd(state, target_full)
        
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
            
    def save(self, filename):
        self.model.save_weights(filename)
        
    def load(self, filename):
        self.model.load_weights(filename)

        

#class for the Q-value model
class LinearModel():
    def __init__(self, input_dim, n_action):
        #initialize model parameters
        self.W = np.random.normal(size = input_dim * n_action) / np.sqrt(input_dim)
        self.W = self.W.reshape(input_dim, n_action)
        self.b = np.zeros(n_action)
        
        #initialize velocity of the momentum gradient
        self.vW = 0
        self.vb = 0
        
        #record losses during training
        self.losses = []
        
        
    def predict(self, X):
        #X is supposed to be a two dimensional array, to be in line with other modules like tensorflow
        assert(len(X.shape) == 2)
        #linear estimate
        return np.matmul(X, self.W) + self.b

    def sgd(self, X, Y, learning_rate = 0.01, momentum = 0.9):
        assert(len(X.shape) == 2)
        #overall number of samples timesteps * predictions (n_actions)
        num_values = np.prod(Y.shape)

        #generate prediction
        Y_hat = self.predict(X)

        #calculate gradients
        gradW = 2 * np.matmul(np.transpose(X),(Y_hat - Y)) / num_values
        gradb = 2 * (Y_hat - Y).sum(axis = 0) / num_values

        #update velocity according to current gradients
        self.vW = momentum * self.vW - learning_rate * gradW
        self.vb = momentum * self.vb - learning_rate * gradb

        #momentum based parameter update
        self.W += self.vW
        self.b += self.vb
        
        #collect losses of the learning step
        mse = np.mean((Y_hat - Y)**2)
        self.losses.append(mse)
        
    def save_weights(self, filename):
        np.savez(filename, W = self.W, b = self.b)
        
    def load_weights(self, filename):
        npz = np.load(filename)
        self.W = npz['W']
        self.b = npz['b']
        
        
#class for the environment
class Environment():
    def __init__(self, data, initial_investment = 20000):
        
        self.stock_price_history = data
        self.n_steps, self.n_stocks = self.stock_price_history.shape
        
        self.initial_investment = initial_investment
        self.current_step = None
        self.stocks_owned = None
        self.stock_prices = None
        self.cash_in_hand = None
        
        self.action_space = np.arange(3**self.n_stocks)
        self.action_list = list(map(list, itertools.product([0,1,2], repeat = self.n_stocks)))
        
        self.state_dim = 2 * self.n_stocks + 1
        
        self.reset()
        
    def reset(self):
        self.current_step = 0
        self.stocks_owned = np.zeros(self.n_stocks)
        self.stock_prices = self.stock_price_history[self.current_step]
        self.cash_in_hand = self.initial_investment
        return self._get_obs()
        
    def step(self, action):
        assert action in self.action_space
        
        #initial value of the portfolio at start
        previous_value = self._get_val()
        
        #lets see what happens in the next step
        self.current_step += 1
        self.stock_prices = self.stock_price_history[self.current_step]
        
        #perform the trade
        self._trade(action)
        
        #get the value after the trade
        current_value = self._get_val()
        
        #reward: increase in the portfolio value
        reward = current_value - previous_value
        
        #are we at the end of the timeline
        done = self.current_step == self.n_steps - 1
        
        #information for the current value
        info = {'current_value': current_value}
        
        #return the next state, reward, if we are finished and what the current value is
        return self._get_obs(), reward, done, info
        
    def _get_val(self):
        return np.matmul(self.stocks_owned, self.stock_prices) + self.cash_in_hand
    
    def _get_obs(self):
        obs = np.empty(self.state_dim)
        obs[:self.n_stocks] = self.stocks_owned
        obs[self.n_stocks:2*self.n_stocks] = self.stock_prices
        obs[-1] = self.cash_in_hand
        return obs
    
    def _trade(self, action):
        
        action_vec = self.action_list[action]
        
        sell_index = []
        buy_index = []
        
        for i, a in enumerate(action_vec):
            if a == 0:
                sell_index.append(i)
            elif a == 2:
                buy_index.append(i)

        if sell_index:
            for s in sell_index:
                self.cash_in_hand += self.stocks_owned[s] * self.stock_prices[s]
                self.stocks_owned[s] = 0

        if buy_index:
        #buy the stocks
            can_buy = True
            while can_buy:
                for b in buy_index:
                    if self.cash_in_hand >= self.stock_prices[b]:
                        self.stocks_owned[b] += 1
                        self.cash_in_hand -= self.stock_prices[b]
                    else:
                        can_buy = False

#function to scale examples of the states 
def get_scaler(environment):
    
    states = []
    done = False
    while not done:
        action = random.choice(environment.action_space)
        next_state, reward, done, info = environment.step(action)
        states.append(next_state)

    states = np.array(states)
    scaler = StandardScaler()
    scaler = scaler.fit(states)
    
    return scaler

#function to play one episode
def play_one_episode(environment, agent, is_train = True):

    state = environment.reset()
    state = scaler.transform([state])
    done = False
 
    while not done:
        action = agent.act(state)
        next_state, reward, done, info = environment.step(action)
        next_state = scaler.transform([next_state])
        if is_train:
            agent.train(state, action, reward, next_state, done)
        state = next_state
        
    return info['current_value']
```

```python
#load the data
data = get_data()

#half is training, half is testing data
n_timesplit = data.shape[0] // 2

#split the data
train_data = data[:n_timesplit]
test_data = data[n_timesplit:]

#train the agent for e episodes
num_episodes = 2000

#initialize the environment with training data
env = Environment(train_data)

#generate the scaler via random play of an episode
scaler = get_scaler(env)
#save the scaler for test data
with open(f'scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

#initialize the agent
agent = DQNAgent(env.state_dim, len(env.action_space))

#save the portfolio value at the end of training data
portfolio_value = []
for e in range(num_episodes):
    t0 = datetime.now()
    val = play_one_episode(env, agent, True)
    portfolio_value.append(val)
    if (e+1) % 100 == 0:
        t1 = datetime.now()
        print(f'The value after {e+1} episodes is: {val:.2f}; this episode took: {t1 - t0}')

agent.save('linear_model_params.npz')
np.save('portfolio_values_training.npy', portfolio_value)
```

```python
pf_values_train = np.load('portfolio_values_training.npy')
plt.plot(pf_values_train)
plt.show()
print(f'average: {np.mean(pf_values_train):.2f}, minimum: {np.min(pf_values_train):.2f}, max: {np.max(pf_values_train):.2f}')
```

```python
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
    
env = Environment(test_data)
agent = DQNAgent(env.state_dim, len(env.action_space))
agent.epsilon = 0.01
agent.load('linear_model_params.npz')

#save the portfolio value at the end of training data
portfolio_value = []
for e in range(num_episodes):
    t0 = datetime.now()
    val = play_one_episode(env, agent, is_train = False)
    portfolio_value.append(val)
    if (e+1) % 100 == 0:
        t1 = datetime.now()
        print(f'The value after {e+1} episodes is: {val:.2f}; this episode took: {t1 - t0}')

#show losses
plt.hist(portfolio_value, bins = 20)
plt.show()
print(f'average: {np.mean(portfolio_value):.2f}, minimum: {np.min(portfolio_value):.2f}, max: {np.max(portfolio_value):.2f}')
```
